## About the dataset

There are `245955 rows` and `23 columns` in the dataset. The data contains information about airline routes in US for the time period of 1993-2024. Each year contains 4 quarters. The attributes of the dataset include the city and airport (id and name) for depature and arrival. The dataset also includes columns `passengers` and `fare` which, according to the description of the dataset on Kaggle, are average number of passengers and average fare amount for each airline route presumabily for each quater (there was no comment about it). It also has brief market related data for each attribute such as the largest and "lowest" carrier (company) with their respective market share. Lastly, it has latitude and longitude for depature and arrival airports. `tbl1apk` column acts as a unique id for every single route of each year and quarter. There are a lot of missing values in a number of columns. We will investigate and take a closer look later.

## Next step in refining the data

There are a few things to be done in the dataset to "clean" it. 

- We should first rename the `passenger` and `fare` columns to reflect that they are average. 
- We will drop the `tbl` column as it provides no information. It is most likely just a column generated by the web scraper or some sort.
- Investigate each column with missing value and refine them. While this is a very generic description of what is needed to be done, exploration of each column would rather be documented when we get to each column as the missing values might not be uniform.
- For Geocode columns, there is some inconsistency in how some rows are documented. Basically, these columns contain latitude and longitude of the airports but the text around the lat and long values can be inconsistent. We may have to use regex to extract them out.
- Upon checking the `nsmiles` column, which contains miles flown by the plane for that particular route, we found that there are some inconsistent values. For example, flight flown from airport id 10135 to 14112 is noted as 970 miles for a lot of the rows but some rows are noted as 971 miles. This is likely to be the inconsistency generated by rounding up/down during the data generation process. We will have to check if there are any other case in the column and fix them. This is the only possible column with such issue since the distance is going to be fixed despite the year or quarter change.
- Similar to the last point, we ought to check if there are any syntactical errors in other columns.

```python
## TODO: Check if there are any routes with ABSOLUTELY missing geocode values

# Finding out which routes have absolutely no geocode values
# Group by departure and arrival airport IDs and filter groups where all geocode values are NaN
consistently_nan_routes = df.groupby(['airportid_1', 'airportid_2']).filter(
    lambda x: x['Geocoded_City1'].isna().all() and x['Geocoded_City2'].isna().all()
)
```

```python
# Find how many unique airline routes have missing geocode values
unique_geo_nan_routes = consistently_nan_routes[['airportid_1', 'airportid_2']].drop_duplicates()
print(unique_geo_nan_routes.shape[0])
```

```python
#create routes column for grouping later
df["route"] = df["airport_1"] + "-" + df["airport_2"]
df.head()
```

```python
unique_routes_depature = df[['route', 'Geocoded_City1']].dropna().drop_duplicates(subset='route')
unique_routes_arrival = df[['route', 'Geocoded_City2']].dropna().drop_duplicates(subset='route')

df = df.merge(unique_routes_depature, on='route', how='left', suffixes=('', '_fixed'))
df = df.merge(unique_routes_arrival, on='route', how='left',  suffixes=('', '_fixed'))
df.drop(columns=['Geocoded_City1', 'Geocoded_City2'], inplace=True)
```

```python
df.head()
```

```python
#remove flights with missing values (334)
df.dropna(subset=['Geocoded_City1_fixed'], inplace=True)
```

```python
#regex to extract coordinates only - standardise
df['coordinates_city1'] = df['Geocoded_City1_fixed'].str.extract(r'(\(\d+.\d+\,\ \-?\d+.\d+\))')
df['coordinates_city2'] = df['Geocoded_City2_fixed'].str.extract(r'(\(\d+.\d+\,\ \-?\d+.\d+\))')
```

```python
df = df.drop('Geocoded_City1_fixed', axis=1)
df = df.drop('Geocoded_City2_fixed', axis=1)
```

```python
df.head()
```

```python
df['airportid-route'] = df[['airportid_1', 'airportid_2']].apply(lambda x: str(x['airportid_1']) + str(x['airportid_2']), axis=1)
```

```python
# Fix distance column
def replace_with_mode(group):
    mode_distance = group['nsmiles'].mode()[0]
    group['nsmiles'] = mode_distance
    return group

# Apply the function to each group
df = df.groupby(['route']).apply(replace_with_mode)
df = df.reset_index(drop=True)
```

```python
# Group by 'route' and aggregate 'nsmiles' to find unique values
grouped = df.groupby('airportid-route')['nsmiles'].unique()

# Find routes with more than one unique mileage
inconsistent_routes = grouped[grouped.apply(len) > 1]

# Count the number of inconsistent routes and rows
num_inconsistent_routes = len(inconsistent_routes)
num_inconsistent_rows = df[df['airportid-route'].isin(inconsistent_routes.index)].shape[0]

print(f"Number of routes with inconsistent mileage: {num_inconsistent_routes}")
print(f"Number of rows with inconsistent mileage: {num_inconsistent_rows}")
```
